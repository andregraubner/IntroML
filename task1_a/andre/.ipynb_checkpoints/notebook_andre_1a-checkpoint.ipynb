```python
# Make imports
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

df_train = pd.read_csv('../Data/train.csv')
```

```python
# Structure data

X = df_train.drop(['Id', 'y'], axis=1)
y = df_train['y']
```

```python
# Plot data

plt.figure(figsize=(20,10))

plt.plot(y.index, y, color='red')

for i in range(1,10):
    plt.plot(X[f'x{i}'].index, X[f'x{i}'], color='blue')
```

```python
# Cross validation

a = [0.1, 1, 10, 100, 1000]

for i in range(1,5):
    model = Ridge(alpha=a[i])
    neg_MSE = cross_val_score(model, X, y, cv=10, scoring='neg_mean_squared_error')
    av_RMSE = np.average(np.sqrt(-neg_MSE))
    print('With alpha=' + str(a[i]) + ': ' + str(av_RMSE))
```
